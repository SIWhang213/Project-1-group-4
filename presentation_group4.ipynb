## How much does US gasolin prices affect the stock market in the US oil industry?
- stock price from yahoo finance
- gasoline price in the US from FRED (https://fred.stlouisfed.org/):  the average price of all unleaded regular gasoline
- line regression and r-value between gasoline price and oil company's stock price

- Exxon Mobil Corporation (XOM), Chevron Corporation(CVX), ConocoPhillips(COP), Marathon Petroleum Corporation(MPC) have strong correlation with gasoline price.
